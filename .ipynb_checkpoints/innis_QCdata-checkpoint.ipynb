## MEAN MONTHLY VALUES
Make list for each month, using 34 years of data

```python
jan = []
feb = []
mar = []
apr = []
may = []
jun = []
jul = []
aug = []
sep = []
octo = []
nov = []
dec = []

for k in range(len(days)):
    dt = netCDF4.num2date(tt[days[k]],u'days since 1985-01-01 00:00:00 UTC')
    if dt.month == 1:
        jan.append(k)
    if dt.month == 2:
        feb.append(k)
    if dt.month == 3:
        mar.append(k)
    if dt.month == 4:
        apr.append(k)
    if dt.month == 5:
        may.append(k)
    if dt.month == 6:
        jun.append(k)
    if dt.month == 7:
        jul.append(k)
    if dt.month == 8:
        aug.append(k)
    if dt.month == 9:
        sep.append(k)
    if dt.month == 10:
        octo.append(k)
    if dt.month == 11:
        nov.append(k)
    if dt.month == 12:
        dec.append(k)
```

GET SEASONAL VALUES

```python
DJF = []
MAM = []
JJA = []
SON = []


for k in range(len(days)):
    dt = netCDF4.num2date(tt[days[k]],u'days since 1985-01-01 00:00:00 UTC')
    if dt.month == 1:
        DJF.append(k)
    if dt.month == 2:
        DJF.append(k)
    if dt.month == 3:
        MAM.append(k)
    if dt.month == 4:
        MAM.append(k)
    if dt.month == 5:
        MAM.append(k)
    if dt.month == 6:
        JJA.append(k)
    if dt.month == 7:
        JJA.append(k)
    if dt.month == 8:
        JJA.append(k)
    if dt.month == 9:
        SON.append(k)
    if dt.month == 10:
        SON.append(k)
    if dt.month == 11:
        SON.append(k)
    if dt.month == 12:
        DJF.append(k)

DJFw = wh[DJF]
MAMw = wh[MAM]
JJAw = wh[JJA]
SONw = wh[SON]



print 'Season    Min             Max             Mean  stdev' 
print 'DJF', DJFw.min(),DJFw.max(),DJFw.mean(), stats.sem(DJFw)
print 'MAM', MAMw.min(),MAMw.max(),MAMw.mean(), stats.sem(MAMw)
print 'JJA', JJAw.min(),JJAw.max(),JJAw.mean(), stats.sem(JJAw)
print 'SON', SONw.min(),SONw.max(),SONw.mean(), stats.sem(SONw)
```
