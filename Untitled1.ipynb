```python
days1 = [] #daystart
for k in range(len(tt1)):
       
    t1 = netCDF4.num2date(tt1[k],u'days since 1985-01-01 00:00:00 UTC')
    if k == 0: #get day 0. Then else: every day after that
        days1.append(0)
        dd = netCDF4.num2date(tt1[k],u'days since 1985-01-01 00:00:00 UTC')
        it = 0
    else:
        if t1.day != dd.day: #if day 1 is not equal to day 2, then append
            #print dd.day,t1.day
            days1.append(k)
            it += 1
            dd = netCDF4.num2date(tt1[k],u'days since 1985-01-01 00:00:00 UTC')
            
            
print 'Number of Satellite Tracks:', len(tt1[days1])
```
